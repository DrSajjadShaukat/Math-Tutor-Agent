### What This Agent Does
An intelligent AI math tutor that:
- Explains math concepts at PhD level
- Has conversation memory
- Uses plain text (no LaTeX symbols)
- Adapts to your mathematical background

### Tech Stack
- Python 3
- Anthropic Claude API (claude-haiku-4-5)
- Conversation memory system

### How To Run
1. pip install anthropic
2. Add your API key
3. Run all cells
4. Ask any math question!

### Example Questions
- "What is gradient descent?"
- "Explain eigenvalues intuitively"
- "What is the Fourier Transform?"

```python
# ============================================
# MATH TUTOR AGENT
# Author: Dr. Sajjad Shaukat
# Tech: Anthropic Claude API + Memory System
# ============================================

!pip install anthropic -q

import warnings
warnings.filterwarnings('ignore')
import anthropic
import os

# Initialize client
client = anthropic.Anthropic(
    api_key=os.environ.get("ANTHROPIC_API_KEY", "YOUR_API_KEY_HERE")
)

# Conversation memory
conversation = []

def chat(user_message):
    """Send message to Math Tutor and get response"""
    conversation.append({
        "role": "user",
        "content": user_message
    })

    response = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=1024,
        system="""You are an expert Math tutor helping Dr. Sajjad,
        a Math PhD learning AI Engineering.
        RULES:
        - Explain using mathematical concepts they already know
        - No LaTeX or $ signs
        - Write powers as x squared not x^2
        - Be precise and clear
        - Relate AI concepts to mathematics""",
        messages=conversation
    )

    reply = response.content[0].text
    conversation.append({
        "role": "assistant",
        "content": reply
    })
    return reply

# ============================================
# TEST THE AGENT
# ============================================

print("=" * 50)
print("Math Tutor Agent Ready!")
print("=" * 50)

print("\nQuestion 1: Gradient Descent")
print(chat("Explain gradient descent using calculus concepts I already know"))

print("\nQuestion 2: Follow up")
print(chat("Can you give me a simple numerical example of what you just explained?"))
```
